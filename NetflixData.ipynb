```python
# Netflix Data Cleaning, Analysis, and Visualization (Improved & Modularized)
# ---------------------------------------------------------------
# This code implements the suggestions for better structure, feature engineering,
# and advanced visualization. Each block is commented for clarity.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

# -------------------------
# 1. Data Loading Function
# -------------------------
def load_data(filepath):
    """Load Netflix data from a CSV file."""
    return pd.read_csv(filepath)

# -------------------------
# 2. Data Cleaning Function
# -------------------------
def clean_data(df):
    """Clean Netflix data: handle missing values, duplicates, and data types."""
    # Drop duplicate rows
    df = df.drop_duplicates()
    
    # Fill missing values in non-critical columns with 'Unknown'
    for col in ['director', 'cast', 'country']:
        if col in df.columns:
            df[col] = df[col].fillna('Unknown')
    
    # Convert 'date_added' to datetime
    df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
    
    # Drop rows with missing critical info (title/type)
    df = df.dropna(subset=['title', 'type'])
    
    # Reset index after cleaning
    df = df.reset_index(drop=True)
    return df

# ---------------------------------------------
# 3. Feature Engineering: Duration and Genres
# ---------------------------------------------
def engineer_features(df):
    """Extract numeric duration and split genres for analysis."""
    # Extract duration in minutes or number of seasons
    df['duration_int'] = df['duration'].str.extract('(\d+)').astype(float)
    df['duration_type'] = df['duration'].str.extract('([a-zA-Z]+)').iloc[:,0]
    
    # Split 'listed_in' into a list of genres
    df['genres'] = df['listed_in'].apply(lambda x: [genre.strip() for genre in x.split(',')])
    return df

# ---------------------------------------------
# 4. Visualization Functions
# ---------------------------------------------
def plot_content_type_distribution(df):
    """Bar plot for Movies vs TV Shows."""
    type_counts = df['type'].value_counts()
    plt.figure(figsize=(6,4))
    sns.barplot(x=type_counts.index, y=type_counts.values, palette='Set2')
    plt.title('Distribution of Content by Type')
    plt.xlabel('Type')
    plt.ylabel('Count')
    plt.show()

def plot_top_genres(df, top_n=10):
    """Bar plot for top N genres."""
    all_genres = sum(df['genres'], [])
    genre_counts = pd.Series(all_genres).value_counts().head(top_n)
    plt.figure(figsize=(8,5))
    sns.barplot(x=genre_counts.values, y=genre_counts.index, palette='Set3')
    plt.title(f'Top {top_n} Genres on Netflix')
    plt.xlabel('Count')
    plt.ylabel('Genre')
    plt.show()

def plot_content_added_over_time(df):
    """Plot count of content added per year."""
    df['year_added'] = df['date_added'].dt.year
    plt.figure(figsize=(10,5))
    sns.countplot(x='year_added', data=df, palette='coolwarm')
    plt.title('Content Added Over Years')
    plt.xlabel('Year')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()

def plot_top_directors(df, top_n=10):
    """Bar plot for top N directors by number of titles."""
    top_directors = df['director'].value_counts().head(top_n)
    plt.figure(figsize=(8,5))
    sns.barplot(x=top_directors.values, y=top_directors.index, palette='Blues_d')
    plt.title(f'Top {top_n} Directors with Most Titles')
    plt.xlabel('Number of Titles')
    plt.ylabel('Director')
    plt.show()

def plot_wordcloud_titles(df):
    """Generate and display a word cloud of movie titles."""
    movie_titles = df[df['type'] == 'Movie']['title']
    wordcloud = WordCloud(width=800, height=400, background_color='black').generate(' '.join(movie_titles))
    plt.figure(figsize=(10,5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title('Word Cloud of Movie Titles')
    plt.show()

# ---------------------------------------------
# 5. Example Usage
# ---------------------------------------------
if __name__ == "__main__":
    # Load data
    data = load_data('netflix1.csv')
    
    # Clean data
    data = clean_data(data)
    
    # Feature engineering
    data = engineer_features(data)
    
    # Visualizations
    plot_content_type_distribution(data)
    plot_top_genres(data)
    plot_content_added_over_time(data)
    plot_top_directors(data)
    plot_wordcloud_titles(data)
```
